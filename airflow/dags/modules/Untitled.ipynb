```python
import pandas as pd
import isd_io
import numpy as np
import datetime
```

```python
file_data = isd_io.read_isd("temp/888890-99999-2022")
df = pd.DataFrame(file_data)
```

```python
df
```

```python
df['count'] = df.groupby('date')['date'].transform('count')
df = df[df['count'] > 3]
```

```python
df = df.drop('count', axis = 1)
```

```python
df = df.groupby(['station_id', 'date']).mean(numeric_only= False).round(2)
df
```

```python
df['sky_condition'] = np.floor(df['sky_condition'])
def get_sky_condition(key):
    mapping = {        
        0: "SKC or CLR",
        1: "One okta",
        2: "Two okta",
        3: "Three okta",
        4: "Four okta",
        5: "Five okta",
        6: "Six okta",
        7: "Seven okta",
        8: "Eight okta",
        9: "Sky obscured, or cloud amount cannot be estimated",
        10: "Partial obscuration",
        11: "Thin scattered",
        12: "Scattered",
        13: "Dark scattered",
        14: "Thin broken",
        15: "Broken",
        16: "Dark broken",
        17: "Thin overcast",
        18: "Overcast",
        19: "Dark overcast"
    }

    return mapping[int(key)]
```

```python
df['sky_condition'] = df['sky_condition'].apply(lambda x: get_sky_condition(x) if pd.notnull(x) else x)
df
```
